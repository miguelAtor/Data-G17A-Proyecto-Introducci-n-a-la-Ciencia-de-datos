Ganancia mínima y máxima (dentro del dataset) por deporte y por país.

```python
por_pais = pd.pivot_table(df1, index='Nationality', columns = 'Sport',values="earnings ($ million)", aggfunc='sum', fill_value=0)
```

```python
ganancias = pd.DataFrame({'Country': por_pais.index})
ganancias['Max Earning Sport'] = por_pais.idxmax(axis=1).to_numpy()
ganancias['Max Earning'] = por_pais.max(axis=1).to_numpy()
ganancias['Min Earning Sport'] =  por_pais.idxmin(axis=1).to_numpy()
ganancias['Min Earning'] = por_pais.min(axis=1).to_numpy()
ganancias.set_index('Country', inplace = True)
ganancias.style.background_gradient(cmap='Reds') 
```

Atleta con mayores ganancias por deporte por década

```python
def def_decade(y):
    if y <= 2000:
        return '90s'
    elif y <= 2010:
        return '2000s'
    else:
        return '10s'

df2 = df1.loc[:, ('Name', 'Sport', 'earnings ($ million)')]
df2['Decade'] = df1['Year'].apply(def_decade)
```

```python
df2 = pd.pivot_table(df2, index=['Sport', 'Decade'], columns = 'Name',values="earnings ($ million)", aggfunc='sum', fill_value=0)
```

```python
atleta_ganancias = pd.DataFrame()
atleta_ganancias.index = df2.index
atleta_ganancias['Name'] = df2.idxmax(axis=1).to_numpy()
atleta_ganancias['Max earnings'] = df2.max(axis=1).to_numpy()
atleta_ganancias.style.background_gradient(cmap='Reds') 
```

Ganancia total por cada deporte por cada año.

```python
df3 = df1.loc[:, ('Sport', 'Year' ,'earnings ($ million)')]
df3 = pd.pivot_table(df3, index=['Sport', 'Year'],values="earnings ($ million)", aggfunc='sum', fill_value=0)
df3.style.background_gradient(cmap='Reds') 
```
